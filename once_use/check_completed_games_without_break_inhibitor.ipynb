```python
import pandas as pd

strcolumn_dict = {"tournamentId": "str", "matchId":"str", "gameId":"str", "patch":"str", "teamcode_blue_on_detail":"str", "teamcode_red_on_detail":"str", "blueteam_code":"str", "redteam_code":"str"}
temp1 = [f"esportsPlayerId_{i}" for i in range(10)]
temp2 = ["str" for i in range(10)]
esportsPlayerId_type_dict = dict(zip(temp1, temp2))
strcolumn_dict.update(esportsPlayerId_type_dict)
df = pd.read_excel("../../data/game_ids.xlsx", index_col=0, dtype= strcolumn_dict)
df.info()
```

```python
last_row_df = pd.read_excel("../../data/last_row_of_collected_datas.xlsx", dtype={"gameId":"str"})
last_row_df.head()
```

```python
merged_df = pd.merge(df,last_row_df, how="left", on="gameId")
merged_df.info(verbose=True, show_counts=True)
```

```python
weird_df = merged_df[((merged_df["blue_inhibitors"] == 0) & (merged_df["winner_side"] == "Blue")) | ((merged_df["red_inhibitors"] == 0) & (merged_df["winner_side"] == "Red"))]
weird_df.head()
```

```python
desired_columns = ["startTime(match)", "tournamentId", "gameId", "winner_side", "blue_inhibitors", "red_inhibitors", "blueteam_code", "redteam_code", "teamcode_blue_on_detail", "teamcode_red_on_detail", "blue_totalGold", "red_totalGold", "duration"]
desired_columns.extend([f"pick_{i}" for i in range(10)])
desired_columns_of_weird_df = weird_df[desired_columns]
desired_columns_of_weird_df.to_excel("../../data/completed_games_without_break_inhibitor.xlsx")
```
