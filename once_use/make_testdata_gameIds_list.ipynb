```python
import pandas as pd
import os
```

```python
df = pd.read_excel("../data/game_ids_for_test.xlsx", dtype={"tournamentId":"str", "matchId":"str", "gameId":"str", "patch":"str"})
```

```python
df
```

```python
os.makedirs('../data/game_ids_for_test', exist_ok=True)
rows_per_piece = 20
num_pieces = len(df) // rows_per_piece + int(len(df) % rows_per_piece != 0)
for i in range(num_pieces):
    start_row = i * rows_per_piece
    end_row = (i + 1) * rows_per_piece
    piece = df.iloc[start_row:end_row]
    piece.to_excel(f'../data/game_ids_for_test/game_id_for_test_{i}.xlsx', index=False)
```
