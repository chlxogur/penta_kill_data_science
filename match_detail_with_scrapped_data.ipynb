```python
import os
import pandas as pd
from tqdm import tqdm, tqdm_pandas
import numpy as np
```

```python
PARTICIPANT_NUMBER_OF_A_TEAM = 5

PROJ_PATH = "/content/drive/MyDrive/penta_kill/data science/"
former = pd.read_excel(PROJ_PATH + "game_ids.xlsx")
crawled = pd.read_excel(PROJ_PATH + "crawling_result_with_codenames.xlsx")
DETAIL_PATH = PROJ_PATH + "collected_game_data/"
file_list = os.listdir(DETAIL_PATH)
pick_from_detail = []
pick_from_scrap = []
former = former.astype({"tournamentId":"str", "matchId":"str", "gameId":"str"})

for i in range(PARTICIPANT_NUMBER_OF_A_TEAM * 2):
  former[f"pick_{i}"] = np.nan

former.head()
```

```python
def addPick(gameId):
  result = []
  file_list = os.listdir(DETAIL_PATH)
  if gameId + ".xlsx" in file_list:
    detail_data = pd.read_excel(DETAIL_PATH + gameId + ".xlsx")
    for i in range(PARTICIPANT_NUMBER_OF_A_TEAM * 2):
      result.append(detail_data.at[0, f"championName_{i}"])
    return result
  elif gameId + "_invalid.xlsx" in file_list:
    pass
  else:
    print(f"file {gameId} doesn't exists!")
  return [np.nan for i in range(10)]
```

```python
print(PROJ_PATH + "game_ids_with_picks.xlsx")
```

```python
#for tournamentId, games_of_a_tournament_df in former.groupby("tournamentId"): # game_ids_with_patch.xlsx 데이터프레임을 tournamentId별로 나눔 -- merge하면 빠를 것 같아 일단 나눠봄, 중복검사는 필요할 듯
tqdm.pandas()
former[["pick_0", "pick_1", "pick_2", "pick_3", "pick_4", "pick_5", "pick_6", "pick_7", "pick_8", "pick_9"]] = former.progress_apply(lambda row : addPick(row["gameId"]), axis = 1)

former.to_excel(PROJ_PATH + "game_ids.xlsx", index = None)
```
